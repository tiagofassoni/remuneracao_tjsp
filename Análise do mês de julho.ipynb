```python
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
base_julho = pd.read_csv('julho.csv',decimal=',', thousands='.' )
```

# Quantos magistrados o TJSP possui?


Primeiro, temos que filtrar quem está lá. 

```python
pessoal_no_tj = base_julho[base_julho['Lotação'].str.contains('TRIBUNAL DE JUSTIÇA')]
```

Depois, agrupar os dados.

## Numéro de magistrados aí


```python
pessoal_no_tj.groupby('Cargo').size()
```

# Qual o Valor Médio do Vencimento?

O vencimento é a remuneração paradigma. E, além do valor médio do vencimento, precisamos do desvio padrão.

```python
ProfileReport(pessoal_no_tj[['Cargo',r'RemuneraçãoParadigma(1)']])
```

## Resultados do valor médio de vencimento

A média é 29.235 reais, mas essa média é puxada para baixo por uns poucos gatos pingados que ganham bem pouco. O salário mais frequente é o máximo, de 30471.11 reais

Outro ponto sensacional é que só há 6 valores diferentes de salário.

# Qual o valor médio pago com verbas indenizatórias?
Verbas indenizatórias = vantagens pessoais + indenizações + vantagens eventuais

Estou pegando somente para junho

```python
indenizatorias = pessoal_no_tj['VantagensPessoais(2)'] + pessoal_no_tj['Indenizações(3)'] + pessoal_no_tj['VantagensEventuais(4)']
# indenizatorias = indenizatorias.rename('VerbasIndenizatori')
pessoal_no_tj['VerbasIndenizatorias'] = indenizatorias

```

```python
ProfileReport(pessoal_no_tj[['Cargo',r'VerbasIndenizatorias']])
```

Eeita, mudou (e para maior) a quantidade de verbas indenizatórias.

## Falta uma questão aí: quais cargos ganham verbas indenizatórias?

```python
pessoal_no_tj[pessoal_no_tj['VerbasIndenizatorias'] > 0].groupby('Cargo').size()
```

Mais desembargadores ganharam verbas indenizatórias que em junho

```python
pessoal_no_tj[pessoal_no_tj['VerbasIndenizatorias'] >= 50000]
```
